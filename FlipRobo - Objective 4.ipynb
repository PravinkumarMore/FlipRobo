```python
# Import essential libraries

import numpy as np
import pandas as pd

from selenium import webdriver
from bs4 import BeautifulSoup

import requests
import urllib.request
from urllib.request import urlopen
```

```python
# Configure webdriver to use Chrome Browser

driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://bookpage.com/reviews")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
name = soup.find_all('h4', class_ = 'italic')

Book_Name = []

for i in name:
    Book_Name.append(i.a.text)
    
len(Book_Name)
Book_Name
```

```python
ath_name = soup.find_all('p', class_ = 'sans bold')

Auther_Name = []

for i in ath_name:
    Auther_Name.append(i.text.replace('\n', ""))
    
len(Auther_Name)
```

```python
genre = soup.find_all('p', class_ = 'genre-links hidden-phone')

Genre = []

for i in genre:
    Genre.append(i.text.replace('\n', ""))
    
Genre
```
