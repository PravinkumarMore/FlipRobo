```python
# Import essential libraries

from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Configure webdriver to use Chrome Browser

driver = webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("http://www.imdb.com/chart/top")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
movie_containers100 = soup.find_all('td', attrs = {'class' : 'titleColumn'})
```

```python
# Lists to store the scraped data in
Name = []
IMDB_ratings = []
Years = []
```

```python
rating = soup.find_all('td', class_ = 'ratingColumn imdbRating')

for i in rating:
    ratings = i.strong.text
    IMDB_ratings.append(ratings)
```

```python
IMDB_ratings
```

```python
for i in movie_containers100:
    name = i.a.text
    Name.append(name)
```

```python
for i in movie_containers100:
    year = i.span.text
    Years.append(year)
```

```python
print(len(Name))
print(len(IMDB_ratings))
print(len(Years))
```

### Step 4 : Store the data in the required format

```python
df = pd.DataFrame({'Name' : Name[0:100], 'Ratings' : IMDB_ratings[0:100], 'Years' : Years[0:100] })
df.to_csv('IMDB’s_Top_rated_100_movies.csv', index = False, encoding = 'utf-8-sig')
```
